## Investigation Overview


> In this analysis, I was most interested to find common patterns that influenced a listing annual percentage rate and its rating. I was also interested in gathering insight from certain attributes in the data.


## Dataset Overview

> The cleaned version of the original Loan Data from Prosper is used with a structure of 84984 listings and 23 attributes.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('ProsperLoanData_Clean.csv')
```

```python
# set base color
base_color = sb.color_palette()[0]
```

## Most Common Listings

The Debt Consolidation category was found to be the most common reason people came to Prosper looking for a loan. Followed by Other, Home Improvement, Business and Auto in their respective order.

```python
# Order by ListingCategories by Occurence and select top 5
lc_ord = df['ListingCategory'].value_counts().sort_values(ascending=False)[:5].index
lc_ord
```

```python
plt.figure(figsize=[8, 5])
sb.countplot(data=df, y='ListingCategory', color=base_color, order=lc_ord)
plt.ylabel('Listing Category')
plt.xlabel('Count')
plt.title('Most Common Listing Categories');
```

## Percentage of Borrowers With Verifiable Income

> Income is one of the major factors indicative of a orrower's capacity to pay back a loan. Interestingly enough, 91.36% of the listings in the dataset had a borrower with verifiable income behind them.

```python
vi_counts = df['IncomeVerifiable'].value_counts()
labels = pd.Series(vi_counts.index).replace([True, False], ['Verfiable', 'Non-Verfiable']).values
plt.pie(vi_counts, labels=labels, autopct='%1.2f%%', shadow=True, startangle=200, explode=(0, 0.3))
plt.title('Percentage Of Borrowers With Verfiable Income');
```

## Borrower APR and Prosper Rating Correlation

> BorrowerAPR and ProsperRating were found to have the highest correlation with a score of -0.96. Where a decrease in the rating resulted in an increase in the annual performace rate, and vice versa.


```python
plt.figure(figsize=[20, 5])

plt.subplot(1, 2, 1)
sb.regplot(data=df, x='ProsperRating_numeric', y='BorrowerAPR', truncate=False, x_jitter=0.3, scatter_kws={'alpha': 1/20})
plt.xlabel('ProsperRating')
plt.ylabel('BorrowerAPR')
plt.title('ProsperRating and BorrowerAPR Correlation')

plt.subplot(1, 2, 2)
bins_x = np.arange(0, df['ProsperRating_numeric'].max()+1, 1)
bins_y = np.arange(0, df['BorrowerAPR'].max()+0.03, 0.03)
plt.hist2d(data=df, x='ProsperRating_numeric', y='BorrowerAPR', bins=[bins_x, bins_y], cmap='viridis_r', cmin=0.03)
plt.colorbar()
plt.xlabel('ProsperRating')
plt.ylabel('BorrowerAPR')
plt.title('ProsperRating and BorrowerAPR Correlation');
```

## Borrower APR and Loan Amount Correlation

> A significant negative correlation between LoanOriginalAmount and BorrowerAPR was also observed. Which can be explained by the need of lenders to get an acceptable return on their investment. The lower the loan amount the higher the rate would have to be for them to get a decent incentive to lend.


```python
plt.figure(figsize=[10, 5])

sb.regplot(data=df, x='LoanOriginalAmount', y='BorrowerAPR', truncate=False, x_jitter=0.3, scatter_kws={'alpha': 1/20})
plt.xlabel('Loan Amount')
plt.ylabel('BorrowerAPR')
plt.title('Loan Amount and BorrowerAPR Correlation');
```

## Impact of Income Range on BorrowerAPR, LoanOriginalAmount and Prosper Rating

> IncomeRange was found to positively impact ProsperRating and Loan Amount of a listing, where higher income ranges benefited higher ratings and amounts. While it negatively correlated with BorrowerAPR, where the BorrowerAPR tended to decrease noticeably as the IncomeRange gets higher.


```python
IncomeRangeCategories = ['$0', 'Not employed', '$1-24,999', '$25,000-49,999', '$50,000-74,999', '$75,000-99,999', '$100,000+']
df['IncomeRange'] = pd.Categorical(df['IncomeRange'], categories=IncomeRangeCategories)
g = sb.PairGrid(data=df.sort_values('IncomeRange'), x_vars = ['BorrowerAPR', 'LoanOriginalAmount', 'ProsperRating_numeric'], 
                               y_vars=['IncomeRange'],
                               height=4, aspect=1)
g.map(sb.boxplot, color='Pink')
plt.suptitle('Impact of Income Range on BorrowerAPR, LoanOriginalAmount and ProsperRating')
plt.subplots_adjust(top=.8);
```
